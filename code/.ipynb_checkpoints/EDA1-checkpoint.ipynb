```python
from sklearn import cross_validation
```

```python
plays_pos = df['Plays'].dropna()
```

```python
skipplay = pandas.concat([skips_pos, plays_pos], axis =1, join = 'inner')
```

```python
skipplay.head()
```

```python
skipplay['Skips'].max()
```

```python
skipplay['Plays'].max()
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(skipplay['Skips'], skipplay['Plays'], test_size = 0.4, random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel = 'linear', C=1).fit(X_train, y_train)
```
