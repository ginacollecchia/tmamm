```python
# it would be interesting to see the "arc" of an album, according
# to the rating of a given track number. even better would be to
# do this only for albums with most or all tracks rated.

# Plot average rating versus track number

# I'm doing this a lot, so:
def nantozero(df):
    w = np.isnan(df)
    df[w] = 0.0
    return df
```

```python
# get ratings
# ratings = db['My Rating']
ratings0 = nantozero(ratings)
# get track numbers
tracknums = db['Track Number']
tracknums0 = nantozero(tracknums)
# get track count
trackcts = db['Track Count'] # the total number of tracks in the data set
# since track count is in the denominator, if it's nan, let's set 
# it to 1:
w1 = np.isnan(trackcts)
# initialize
trackcts0 = trackcts
trackcts0[w1] = 1.0
```

```python
# we also need to handle the case where the track count is > than a 
# given track number, or is unreasonably high (>40), by excluding it 
# from the data set by setting it to 0. ideally, we would figure out
# what the track count should actually be, but this approach also 
# seems flawed (user might not have the whole album).
TN = len(tracknums0)
for i in xrange(0, TN):
    if (tracknums0[i] > trackcts0[i]) or (trackcts0[i] > 40):
        tracknums0[i] = 0.0
```

```python
# get average rating: find out highest track number, count tracks
# with given track number. only consider ratings that are non-zero.

# by track number
goodtracknums = []
# by the ratio of the track number to length of the album
# (probably not what we want)
goodtracknumsratio = []
# then only add those ratings
goodratings = []
for i in xrange(0, TN):
    # not interested in tracks with a 0 as track number or 
    # where the total track count is only 1
    if (tracknums0[i] > 0.0) and (trackcts0[i] > 1.0) and (ratings[i] > 0.0): 
        goodtracknums.append(tracknums0[i])
        goodratings.append(ratings[i])
        goodtracknumsratio.append(tracknums0[i]/trackcts0[i])
```

```python
# pretty decent sample size
len(goodtracknums)
```

```python
# now, count up how many ratings exist with a given track number
# so we can compute the average rating for a given track number
TNmax = int(max(goodtracknums))
tnsum = []
for i in xrange(1, TNmax+1): # 33 buckets of track numbers here
    tnsum.append(goodtracknums.count(i))
```

```python
# this would be 208 if we hadn't applied a cut off of 40 to the 
# track count
TNmax
```

```python
# now, finally ready to compute the average rating
avgrating = [0.0]*TNmax # not to be confused with "aggrovating"
for i in xrange(0, len(goodratings)):
    for j in xrange(0, TNmax):
        if goodtracknums[i] == float(j+1):
            avgrating[j] += goodratings[i]/tnsum[j]
```

```python
min(tnsum)
```

```python
# so, we just decided by 0, since there existed a number between 0
# and 33 that was not associated with a given track... but python
# knew to set it to 0. thanks python.
min(avgrating)
```

```python
# now we're ready to plot!
plt.plot(avgrating, 'ro')
plt.axis([0, TNmax, 0, 100])
plt.xlabel('Track Number')
plt.ylabel('Average Rating')
plt.show()
# eh. all that for something not that interesting.
```

```python
# is ratio any more interesting?
plt.plot(goodtracknumsratio, goodratings, 'ro')
plt.axis([0, 1, 0, 100])
plt.xlabel('Track Number / Track Count')
plt.ylabel('Average Rating')
plt.show()
# intersting periodicity seems to be happening, but not the best 
# way to visualize what's going on. maybe cloud or cluster graph.
```

```python
## playing with some class stuff...

# create Song class
class Song(object):
    '''song, an entity of album and artist'''
    def __init__(self, name, track_num, play_count, skip_count):
        self.name = name
        self.track_num = track_num
        self.play_count = play_count
        self.skip_count = skip_count
        
# create Album parent class
class Album:
    def __init__(self, name, year, track_total, song):
        self.name = name
        self.year = year
        self.track_total = track_total
        self.song = song
    
# create the parent parent Artist class
class Artist:
    def __init__(self, name, album):
        self.name = name
        self.album = album
```

```python
# i now regret pluralizing everything...
songs = db['Name']
albums = db['Album']
artists = db['Artist']
plays = db['Plays']
skips = db['Skips']
years = db['Year']
# tracknums = db['Track Number']
# trackcts = db['Track Count']
ar = [0] * len(artists)
for i in range(0, len(songs)):
    song = Song(songs[i], tracknums[i], plays[i], skips[i])
    album = Album(albums[i], years[i], trackcts[i], song)
    ar[i] = Artist(artists[i], album)

ar[7].album.name
```

```python
# now let's figure out how many albums an artist has.
ar_sorted = sorted(ar, key=lambda artist: artist.name)
album_dict = {}
last = ar_sorted[0].name
albums_by_artist = ar_sorted[0].album.name
for i in xrange(1, len(ar_sorted)):
    if ar_sorted[i].name == last:
        albums_by_artist.add(ar_sorted[i].album.name) # since it's a set, it won't add duplicates
    else:
        # store in dictionary
        album_dict[last] = albums_by_artist
        # reset to empty set
        albums_by_artist = set()
        last = ar_sorted[i].name
```

```python
# algorithm for determining how many duplicates of a given item exist
def getNumEntities(a):
    b = sorted(a)
    last = b[0]
    first = 0
    count = []
    field_name = []
    for i in xrange(1, len(b)):
        if a[i] == last:
            continue
        else:
            count.append(i - first)
            field_name.append(last)
            last = b[i]
            first = i
    return field_name, count
```

```python
alphabetized_artists = sorted(db['Artist'])
(unique_artists, counts) = getNumEntities(alphabetized_artists)
# now make it a dictionary
num_songs_by_artist = dict(zip(unique_artists, counts))
num_songs_by_artist['Al Green']
```
